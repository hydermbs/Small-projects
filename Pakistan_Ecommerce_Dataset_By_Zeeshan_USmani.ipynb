```python
#find Null in Columns
print(df.isna().sum(axis = 0))
```

```python
#Droped Complete null rows.
df.dropna(how='all', inplace = True)
df
```

```python
#dropped Complete null columns
df.dropna(axis = 1, how ='all',inplace = True)
df
```

```python
#filled null places in data
df['status'].fillna('completed', inplace = True)
df['sku'].fillna('dummy', inplace = True)
df['category_name_1'].fillna('Test', inplace = True)
df['sales_commission_code'].fillna('3025', inplace = True)
df['Customer Since'].fillna('2018-8', inplace = True)
print('Updated Dataframe:')
df
```

```python
#Extract names of the category to find most selling category
df['category_name_1'].drop_duplicates()
```

```python
#Make a group to find most selling category. Which is Mobiles & Tablets
df_group_cat = df.groupby(["category_name_1"])['grand_total','qty_ordered'].sum().sort_values('grand_total',ascending=False) .astype(int)
df_group_cat.style.set_properties(subset=['grand_total'], **{'width': '300px'})
```

```python
#Make a group of Mobiles & Tablets category to find the orders status of them. Cancelled orders are more than complete orders.
group_df = df.groupby(["category_name_1",'status'])['grand_total','qty_ordered'].sum().sort_values('grand_total',ascending=False) .astype(int)
df_group_mob = group_df.loc['Mobiles & Tablets']
df_group_mob.style.set_properties(subset=['grand_total'], **{'width': '300px'})
```

```python
#Make a group to find most selling mobile  by order qty
group = df.groupby(["category_name_1",'sku'])['qty_ordered','grand_total'].sum().sort_values('qty_ordered',ascending=False)
group.loc['Mobiles & Tablets']
```

```python
#Here is a group of total order status
df_order_status = df.groupby(['status'])['qty_ordered','grand_total'].sum().sort_values('qty_ordered',ascending=False) .astype(int)
df_order_status.style.set_properties(subset=['grand_total'], **{'width': '300px'})
```

```python
#Here is a small summary of Mobiles & Tablets Category with there status, Sku, price and count

df_final = df.groupby(['category_name_1','status','sku'])['grand_total'].agg(['sum','count']) .sort_values('count',ascending=False)
df_final.loc['Mobiles & Tablets']
```
